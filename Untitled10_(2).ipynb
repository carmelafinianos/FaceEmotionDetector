```python
import PyPDF2

def extract_text_from_pdf(pdf_path):
    text = ""
    try:
        with open(pdf_path, 'rb') as file:
            reader = PyPDF2.PdfReader(file)
            for page_num in range(len(reader.pages)):
                page = reader.pages[page_num]
                text += page.extract_text()
    except Exception as e:
        print(f"Error reading PDF file: {e}")
    return text

# Example usage
pdf_path = '/content/Environmental Factors.pdf'
extracted_text = extract_text_from_pdf(pdf_path)
print(extracted_text)
```

```python
import re

def preprocess_text(text):
    # Remove non-alphanumeric characters (except spaces)
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    # Convert text to lowercase
    text = text.lower()
    # Replace multiple spaces with a single space
    text = re.sub(r'\s+', ' ', text).strip()
    return text

# Example usage
preprocessed_text = preprocess_text(extracted_text)
```

```python
from transformers import AutoTokenizer, LlamaForCausalLM, TrainingArguments, Trainer, pipeline
from accelerate import Accelerator
from transformers import BitsAndBytesConfig
from datasets import Dataset
import torch

# Split the text into smaller parts for better handling
# Split the text into smaller parts for better handling
texts = [preprocessed_text[i:i+1024] for i in range(0, len(preprocessed_text), 1024)]

# Create a Hugging Face dataset
dataset = Dataset.from_dict({"text": texts})

# Initialize the tokenizer
model_id = "meta-llama/Meta-Llama-3-8B"
tokenizer = AutoTokenizer.from_pretrained(model_id)

# Add a padding token
tokenizer.pad_token = tokenizer.eos_token

# Tokenize the dataset
def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True, max_length=1024)

tokenized_datasets = dataset.map(tokenize_function, batched=True)
print(tokenized_datasets)
```

```python
import transformers
import torch

model_id = "meta-llama/Meta-Llama-3-8B"

pipeline = transformers.pipeline(
    "text-generation", model=model_id, model_kwargs={"torch_dtype": torch.bfloat16}, device_map="auto"
)
# Define a function to get answers from the text
def generate_text(prompt, context):
    input_text = f"{prompt}\n\nContext: {context}\n\nResponse:"
    result = pipeline(input_text, num_return_sequences=1, max_new_tokens=50) # Use max_new_tokens instead of max_length
    return result[0]['generated_text'].split("Response:")[1].strip()


# Example questions
question_2 = "What are the environmental challenges faced by Egypt?"

# Get answers from the texts
answer_2 = generate_text(question_2, preprocessed_text)

print("Answer 2:", answer_2)
```

```python

# Example questions
question = "Key KPIs in the document"

# Get answers from the texts
answer =generate_text(question, preprocessed_text)

print("Answer :", answer)
```
